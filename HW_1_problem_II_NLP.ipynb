```python
'''Finding root by Contraction Mapping'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

def h(x):
  return x**3+2*x**2-7*x-1

def f(x):
  return (1/7)*(x**3+2*x**2-1)

def df(x): #derivative of f
  return (1/7)*(3*x**2+4*x)

def g(x):
  return np.sqrt((1/2)*(-x**3+7*x+1))

def dg(x): #derivative of g
  return (7-3*x**2)/(2*np.sqrt(-2*x**3+14*x+2))
```

```python
#a,b are end-points and dx is the spacing

#plot of h

def H(a,b,dx):
  x=np.arange(a,b,dx) #space points
  v=np.vectorize(h)
  y=v(x) #function values
  Xaxis=np.zeros(len(x)) #x-axis
  plt.plot(x,y,color='b')
  plt.plot(x,Xaxis,color='r')
  plt.xlabel('x-axis')
  plt.ylabel('h(x)')

```

```python

H(-1,5,0.1)

#There is a root between -1 and 0 and one just to left of 2
```

```python
#plot of derivative of f

def DF(a,b,dx):
  x=np.arange(a,b,dx) #space points
  v=np.vectorize(df)
  y=v(x) #function values
  Xaxis=np.zeros(len(x)) #x-axis
  plt.plot(x,y,color='b')
  plt.plot(x,Xaxis,color='r')
  plt.xlabel('x-axis')
  plt.ylabel('df(x)')



```

```python
DF(-1,1,0.1)

#Between -1 and 1 ,derivative is less than 1. So ,by mean value f is a contraction.
```

```python
#plot of derivative of g

def DG(a,b,dx):
  x=np.arange(a,b,dx) #space points
  v=np.vectorize(dg)
  y=v(x) #function values
  Xaxis=np.zeros(len(x)) #x-axis
  plt.plot(x,y,color='b')
  plt.plot(x,Xaxis,color='r')
  plt.xlabel('x-axis')
  plt.ylabel('dg(x)')


```

```python
DG(1.5,2,0.1)

#hence in [1.5,2] contraction mapping can be used to get the second root
```

```python
#function to find the number of decimal places in a float a

def ndecimals(a):
  return str(a)[::-1].find('.')

n=ndecimals(-0.0123)
print(n)
print(type(n))
```

```python
#Contraction mapping to find a root to n decimal places for a valid guess c and given function T

def contraction(c,T):
  root_old=c #initial guess
  root_new=T(c) #first iterate
  error=np.abs(root_old-root_new) #checking convergence by error
  count=1 #couting number of iteration
  while error>10**(-5) :
    root_old=root_new
    root_new=T(root_new)
    error=np.abs(root_old-root_new)
    count=count+1
  else:
    return round(root_new,4),count

#plotting

def plotcontraction(c,T):
  a,b=contraction(c,T)
  x=np.zeros(b)
  x[0]=c
  for j in range(1,b):
    x[j]=T(x[j-1])
  v=np.vectorize(h)
  y=v(x)
  plt.scatter(x,y)
  plt.xlabel('x-axis')
  plt.ylabel('h(x)')

```

```python
#root between -1 and 0 of h

contraction(-0.5,f)
```

```python
#plot of convergence of root by applying f

plotcontraction(-0.5,f)
```

```python
#root before 2 by applying g

contraction(1.5,g)
```

```python
#plot of convergence

plotcontraction(1.5,g)
```
